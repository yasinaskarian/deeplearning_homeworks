```python
import tensorflow as tf
print(tf.__version__)
import matplotlib.pyplot as plt
import numpy as np
from scipy.sparse import coo_matrix
from sklearn.utils import shuffle
```

```python
import sys
import pickle 
import gzip
f = gzip.open('./desktop/mnist.pkl.gz', 'rb')
if sys.version_info < (3,):
    data = pickle.load(f)
else:
    data = pickle.load(f, encoding='bytes')
f.close()
```

```python
(training_images_, training_labels_), (test_images_, test_labels_) = data
```

```python
import matplotlib.pyplot as plt
plt.imshow(training_images_[42])
print(training_labels_[42])
print(training_images_[42])
```

```python
training_images_inverse  = training_images_ - 255.0
test_images_inverse = test_images_ - 255.0
```

```python
training_images = np.append(training_images_,training_images_inverse, axis=0)
test_images = np.append(test_images_,test_images_inverse, axis=0)
```

```python
test_images.shape
```

```python
training_labels = np.append(training_labels_,training_labels_, axis=0)
test_labels = np.append(test_labels_,test_labels_, axis=0)
```

```python
training_images, training_labels = shuffle(training_images, training_labels, random_state=42)
test_images, test_labels = shuffle(test_images, test_labels, random_state=42)
```

```python
import matplotlib.pyplot as plt
plt.imshow(training_images[21])
print(training_labels[21])
print(training_images[21])
```

```python
len(test_labels)
```

```python
training_images = np.reshape(training_images, (120000, 28, 28,1))
test_images = np.reshape(test_images, (20000, 28, 28,1))
```

```python
training_images.shape
```

```python
training_images  = training_images / 255.0
test_images = test_images / 255.0
```

```python
training_images[20]
```

```python
training_labels = tf.keras.utils.to_categorical(training_labels)
test_labels = tf.keras.utils.to_categorical(test_labels)
```

```python
training_labels[20]
```

```python
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.layers import PReLU
```

```python
model = tf.keras.models.Sequential([
                                    Conv2D(8, (7, 7), activation=PReLU(shared_axes=[1, 2])), 
                                    Conv2D(8, (5, 5), activation=PReLU(shared_axes=[1, 2])),
                                    tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dense(10,activation='softmax')
])
```

```python
model.build((None,28,28,1))
print(model.summary())
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss = 'categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(training_images, training_labels,validation_split=0.2, epochs=5, batch_size=128)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
```
